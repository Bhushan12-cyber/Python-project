# Basics of Probability


Question 1. (a)

```python
import random
heads = tails = 0
for _ in range(10000):
    if random.choice(['H','T']) == 'H':
        heads += 1
    else:
        tails += 1
print("Heads : ",heads)
print("Tails : ",tails)
```

Question 1. (b)

```python
sum = 0
for _ in range(10000):
    if random.randint(1,6) + random.randint(1,6) == 7:
        sum += 1;
print("Probabilit of sum 7 :", sum/10000)
```

Question 2. 

```python
def prob_for_six(trials):
    count = 0
    for _ in range(trials):
        if 6 in [random.randint(1,6) for _ in range(10)]:
            count+=1
    return count / trials

print("Probability of getting six : ", prob_for_six(10000))
```

# Conditional Probability and Bayes Theorem

Question 3. (a)

```python
color = ['red']*5 + ['green']*7 + ['blue']*8
cnt_bl = 0
red_given_bl = 0
prev_color = None
for _ in range(1000):
    current = random.choice(color)
    if prev_color == 'blue':
        cnt_bl += 1
    if current == 'red':
        red_given_bl += 1
    prev_color = current
print("probability of drawing a red ball given that the previous ball was blue is: ", red_given_bl / cnt_bl)
```

# Random Variables and Discrete Probability


```python
import numpy as np
values = [1,2,3]
probs = [0.25,0.35,0.4]
sample = np.random.choice(values , size = 1000 , p = probs)
print("Mean: ",np.mean(sample))
print("Variance: ", np.var(sample))
print("standard deviation: ", np.std(sample))
```

# Continuous Random Variables

```python
import matplotlib.pyplot as plt
import seaborn as sns
data = np.random.exponential(scale = 5 , size = 2000)
sns.histplot(data, bins = 30 , kde = True)
plt.title("Exponential distribution")
plt.xlabel("Val")
plt.ylabel("Freq")
```

# Central Limit Theorem

```python
uniform_d = np.random.uniform(0,1,10000)
sample_mean = []
for _ in range(1000):
    sample = np.random.choice(uniform_d, size = 30)
    sample_mean.append(np.mean(sample))
plt .subplot(1,2,1)
sns.histplot(uniform_d , bins=30, kde =True)
plt.title("Uniform distribution")
plt .subplot(1,2,2)
sns.histplot(sample_mean , bins=30, kde =True)
plt.title("Sample mean distribution")    
```
